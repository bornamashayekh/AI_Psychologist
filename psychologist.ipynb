```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

# بارگذاری داده‌ها
df = pd.read_csv('dep.csv')

# نمایش نمودار همبستگی
plt.figure(figsize=(9, 9))
sns.heatmap(df.corr(), annot=True)

# تقسیم داده‌ها به ویژگی‌ها و برچسب‌ها
X = df.drop("dep", axis=1)
y = df['dep']

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X = scaler.fit_transform(X)

# تقسیم داده‌ها به مجموعه‌های آموزشی و تست
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ساخت مدل شبکه عصبی بهبود یافته
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.3))  # Dropout برای جلوگیری از overfitting
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))  # Dropout برای جلوگیری از overfitting
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

# کامپایل کردن مدل
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer='adam', metrics=['accuracy'])

# تنظیم مقدار یادگیری و توقف زودهنگام
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# آموزش مدل
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), 
                    epochs=200, batch_size=32, 
                    callbacks=[reduce_lr, early_stopping])

# ارزیابی دقت مدل
train_accuracy = model.evaluate(X_train, y_train)[1]
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")

test_accuracy = model.evaluate(X_test, y_test)[1]
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")

# پیش‌بینی برای یک نمونه
sample_input = np.array([[1, 3, 3, 33, 0, 0, 2, 1, 2, 2, 2]])
sample_input_scaled = scaler.transform(sample_input)  # نرمال‌سازی داده ورودی
predicted_output = model.predict(sample_input_scaled)[0]

# نمایش نتایج پیش‌بینی
categories = ['low', 'medium', 'high']
plt.bar(categories, predicted_output)
plt.show()
```
